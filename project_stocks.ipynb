```python
import datetime as dt
import pandas as pd
import pandas_datareader.data as web
```

pandas_datareader.data library allows you to get API from yahoo finance

```python
start = dt.datetime(2014,10,7)
end = dt.datetime(2019, 10, 7)
```

With 'start' and 'end' variables you can begin quering the desired stock and it's date range

```python
tsla = web.DataReader('TSLA', 'yahoo', start, end)
abc = web.DataReader('ABC', 'yahoo', start, end)
aapl = web.DataReader('AAPL', 'yahoo', start, end)
amzn = web.DataReader('AMZN', 'yahoo', start, end)
cvs = web.DataReader('CVS', 'yahoo', start, end)
fb = web.DataReader('FB', 'yahoo', start, end)
mck = web.DataReader('MCK', 'yahoo', start, end)
msft = web.DataReader('MSFT', 'yahoo', start, end)
nflx = web.DataReader('NFLX', 'yahoo', start, end)
t = web.DataReader('TS', 'yahoo', start, end)
unh = web.DataReader('UNH', 'yahoo', start, end)
wmt = web.DataReader('WMT', 'yahoo', start, end)
xom = web.DataReader('XOM', 'yahoo', start, end)
ci = web.DataReader('CI', 'yahoo', start, end)
goog = web.DataReader('GOOG', 'yahoo', start, end)
ba = web.DataReader('BA', 'yahoo', start, end)
cnc = web.DataReader('CNC', 'yahoo', start, end)
pfe = web.DataReader('PFE', 'yahoo', start, end)
hum = web.DataReader('HUM', 'yahoo', start, end)
brk_b = web.DataReader('BRK-B', 'yahoo', start, end)
```

Added stock column with each stock's name to their appropriate dataframe. This is to allow Tableau to display each stock's name once database is imported to Tableau.

```python
aapl['stock']='Apple (AAPL)'
abc['stock']='AmerisourceBergen (ABC)'
amzn['stock']='Amazon (AMZN)'
brk_b['stock']='Berkshire Hathaway (BRK.A)'
cvs['stock']='CVS Health (CVS)'
fb['stock']='Facebook (FB)'
mck['stock']='McKesson (MCK)'
msft['stock']='Microsoft (MSFT)'
nflx['stock']='Netflix (NFLX)'
t['stock']='AT&T (T)'
tsla['stock']='Tesla (TSLA)'
unh['stock']='UnitedHealth (UNH)'
wmt['stock']='Walmart (WMT)'
xom['stock']='Exxon Mobil (XOM)'
ci['stock']='Cigna (CI)'
goog['stock']='Google (GOOG)'
ba['stock']='Boeing (BA)'
cnc['stock']='Centene (CNC)'
pfe['stock']='Pfizer (PFE)'
hum['stock']='Humana (HUM)'
```

```python
aapl = aapl[['stock'] + aapl.columns[:-1].tolist()]
abc = abc[['stock'] + abc.columns[:-1].tolist()]
amzn = amzn[['stock'] + amzn.columns[:-1].tolist()]
brk_b = brk_b[['stock'] + brk_b.columns[:-1].tolist()]
cvs = cvs[['stock'] + cvs.columns[:-1].tolist()]
fb = fb[['stock'] + fb.columns[:-1].tolist()]
mck = mck[['stock'] + mck.columns[:-1].tolist()]
msft = msft[['stock'] + msft.columns[:-1].tolist()]
nflx = nflx[['stock'] + nflx.columns[:-1].tolist()]
t = t[['stock'] + t.columns[:-1].tolist()]
tsla = tsla[['stock'] + tsla.columns[:-1].tolist()]
unh = unh[['stock'] + unh.columns[:-1].tolist()]
wmt = wmt[['stock'] + wmt.columns[:-1].tolist()]
xom = xom[['stock'] + xom.columns[:-1].tolist()]
ci = ci[['stock'] + ci.columns[:-1].tolist()]
goog = goog[['stock'] + goog.columns[:-1].tolist()]
ba = ba[['stock'] + ba.columns[:-1].tolist()]
cnc = cnc[['stock'] + cnc.columns[:-1].tolist()]
pfe = pfe[['stock'] + pfe.columns[:-1].tolist()]
hum = hum[['stock'] + hum.columns[:-1].tolist()]
```

All stock dataframes combined into list and concat together.

```python
stock_frames=[aapl, abc, amzn, brk_b, cvs, fb, mck, msft, nflx, t, tsla, unh, wmt, xom, ci, goog, ba, cnc, pfe, hum]
```

```python
combo_frames = pd.concat(stock_frames)
```

Combined stock dataframe saved as csv database for Tableau to display visualization.
